```python
def plot_feature_distributions(
    df: pd.DataFrame,
    features: Iterable[str],
    bins: int = 30,
    figsize: tuple[int, int] = (20, 15)
) -> None:
    """
    Menampilkan histogram distribusi untuk beberapa fitur numerik
    dalam sebuah DataFrame.

    Parameter
    ----------
    df : pd.DataFrame
        DataFrame yang berisi data.
    features : Iterable[str]
        Daftar nama kolom numerik yang ingin divisualisasikan.
    bins : int, default=30
        Jumlah bin yang digunakan pada setiap histogram.
    figsize : tuple[int, int], default=(20, 15)
        Ukuran figure matplotlib dalam satuan inci (width, height).

    Returns
    -------
    None
    """

    n_features = len(features)
    n_cols = 4
    n_rows = (n_features + n_cols - 1) // n_cols

    plt.figure(figsize=figsize)

    for i, feature in enumerate(features, 1):
        plt.subplot(n_rows, n_cols, i)
        plt.hist(df[feature].dropna(), bins=bins)
        plt.title(f"Distribution of {feature}")
        plt.xlabel(feature)
        plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()
```

#### Pre-Processing Dataset

Drop Target Tidak Perlu

```python
def pre_processing_klasifikasi(
    df: pd.DataFrame,
    columns_to_drop: Iterable[str] = (
        "Sisa_Usia_Jam",
        "Hour",
        "Source_Image",
        "Augment_ID",
    )
) -> pd.DataFrame:
    """
    Melakukan preprocessing awal untuk data klasifikasi dengan
    menghapus kolom-kolom tertentu jika tersedia pada DataFrame.

    Parameter
    ----------
    df : pd.DataFrame
        DataFrame yang akan diproses.
    columns_to_drop : Iterable[str], optional
        Daftar nama kolom yang ingin dihapus dari DataFrame.
        Kolom hanya akan dihapus jika terdapat dalam DataFrame.

    Returns
    -------
    pd.DataFrame
        DataFrame hasil preprocessing dengan kolom yang ditentukan
        telah dihapus (jika tersedia). DataFrame asli tidak dimodifikasi
        karena dilakukan proses copy terlebih dahulu.
    """

    processed_df = df.copy()

    # Hanya drop kolom yang tersedia
    existing_cols = [col for col in columns_to_drop if col in processed_df.columns]

    # Drop kolom yang tersedia
    if existing_cols:
        processed_df = processed_df.drop(columns=existing_cols)

    return processed_df
```

Manual Encoding Target

```python
def manual_encoding_target(df: pd.DataFrame) -> pd.DataFrame:
    """
    Melakukan encoding manual pada kolom target "Kesegaran"
    dengan mengubah label kategorikal menjadi nilai numerik.

    Mapping yang digunakan
    ----------------------
    - "Segar"        -> 0
    - "Kurang Segar" -> 1
    - "Tidak Segar"  -> 2

    Parameter
    ----------
    df : pd.DataFrame
        DataFrame yang berisi kolom target "Kesegaran"
        yang akan dilakukan encoding.

    Returns
    -------
    pd.DataFrame
        DataFrame hasil encoding dengan kolom "Kesegaran"
        dalam bentuk numerik. DataFrame asli tidak dimodifikasi
        karena dilakukan proses copy terlebih dahulu.
    """

    df_copy = df.copy()

    mapping_kesegaran = {
        "Segar": 0,
        "Kurang Segar": 1,
        "Tidak Segar": 2
    }

    df_copy["Kesegaran"] = df_copy["Kesegaran"].map(mapping_kesegaran)

    return df_copy
```

```python
# Proses Manual Encoding
df_train_cls = manual_encoding_target(df_train_cls)
df_test_cls = manual_encoding_target(df_test_cls)
```

#### Training Model Baseline

Fungsi Training dan Evaluasi Baseline Model

```python
def train_baseline_models(
    X_train: pd.DataFrame,
    y_train: pd.Series,
    models: dict,
    k_folds: int = 10,
    random_state: int = 42
):
    """
    Melatih dan mengevaluasi beberapa model baseline klasifikasi menggunakan
    Stratified K-Fold cross-validation, lalu menyimpan artefak evaluasi
    serta model final dalam format ONNX dan PKL.

    Untuk setiap model pada variabel `models`, fungsi ini akan:
    1) Membuat pipeline sederhana: Pipeline([("model", estimator)]).
    2) Melakukan prediksi cross-validation dengan `cross_val_predict`.
    3) Menghitung metrik klasifikasi: Accuracy, Precision/Recall/F1 (weighted).
    4) Membuat confusion matrix dan menyimpannya sebagai file gambar (.jpg).
    5) Menyimpan classification report per model ke file CSV.
    6) Melatih model final pada seluruh data training.
    7) Mengekspor model final ke format ONNX (target_opset=12).
    8) Menyimpan model final dalam format PKL menggunakan joblib.

    Parameter
    ----------
    X_train : pd.DataFrame
        Fitur data training berukuran (n_samples, n_features).
        Dengan semua fitur berupa numerik karena ekspor ONNX menggunakan
        FloatTensorType dan input bertipe float.
    y_train : pd.Series
        Target/label data training berukuran (n_samples,).
    models : dict
        Dictionary berisi pasangan {nama_model: estimator_sklearn}.
        Contoh: {"RF": RandomForestClassifier(...), "SVC": SVC(...)}.
    k_folds : int, default=10
        Jumlah fold untuk Stratified K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada StratifiedKFold (shuffle=True).

    Returns
    -------
    pd.DataFrame
        DataFrame ringkasan performa setiap model dengan kolom:
        ["Model", "Accuracy", "Precision_weighted", "Recall_weighted", "F1_weighted"].
        Ringkasan diurutkan menurun berdasarkan Accuracy dan juga disimpan ke CSV.

    Side Effects
    ------------
    - Menyimpan confusion matrix per model:
      f"{CONF_MATRIX_DIR}/Confusion_Matrix_{model_name}.jpg"
    - Menyimpan classification report per model:
      f"{CONF_MATRIX_DIR}/Metric_Evaluation_{model_name}.csv"
    - Menyimpan model final per model:
      ONNX: f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.onnx"
      PKL : f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.pkl"
    - Menyimpan ringkasan seluruh model:
      f"{ROOT_FOLDER}/Klasifikasi/Baseline_Report/Baseline_Summary_Report.csv"
    """

    cv = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=random_state)

    summary_rows = []

    for model_name, estimator in models.items():

        pipe = Pipeline([("model", estimator)])

        # Cross Validation Metrics
        y_pred_cv = cross_val_predict(pipe, X_train, y_train, cv=cv, n_jobs=-1)

        acc = accuracy_score(y_train, y_pred_cv)
        prec = precision_score(y_train, y_pred_cv, average="weighted")
        rec = recall_score(y_train, y_pred_cv, average="weighted")
        f1 = f1_score(y_train, y_pred_cv, average="weighted")

        summary_rows.append({
            "Model": model_name,
            "Accuracy": acc,
            "Precision_weighted": prec,
            "Recall_weighted": rec,
            "F1_weighted": f1
        })

        # Confusion Matrix
        cm = confusion_matrix(y_train, y_pred_cv)

        plt.figure(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"Confusion Matrix - {model_name}")
        plt.ylabel("Actual")
        plt.xlabel("Predicted")

        cm_path = f"{CONF_MATRIX_DIR}/Confusion_Matrix_{model_name}.jpg"
        plt.tight_layout()
        plt.savefig(cm_path)
        plt.close()

        # Metric CSV per model
        metric_df = pd.DataFrame(classification_report(
            y_train,
            y_pred_cv,
            output_dict=True
        )).transpose()

        metric_csv_path = f"{CONF_MATRIX_DIR}/Metric_Evaluation_{model_name}.csv"
        metric_df.to_csv(metric_csv_path)

        # Fit Final Model
        pipe.fit(X_train, y_train)

        # Save ONNX (Android)
        initial_type = [("float_input", FloatTensorType([None, X_train.shape[1]]))]
        onnx_model = convert_sklearn(pipe, initial_types=initial_type, target_opset=12)

        onnx_path = f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.onnx"
        with open(onnx_path, "wb") as f:
            f.write(onnx_model.SerializeToString())

        # Save PKL (Python)
        pkl_path = f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.pkl"
        joblib.dump(pipe, pkl_path)

        print(f"[DONE] {model_name} selesai disimpan.")

    # Summary CSV semua model
    summary_df = pd.DataFrame(summary_rows).sort_values(by="Accuracy", ascending=False)
    summary_df.to_csv(f"{ROOT_FOLDER}/Klasifikasi/Baseline_Report/Baseline_Summary_Report.csv", index=False)

    return summary_df
```

```python
# Target Folder Penyimpanan hasil
BASELINE_MODEL_DIR = f"{ROOT_FOLDER}/Klasifikasi/Baseline_Report/Baseline_Model"
CONF_MATRIX_DIR = f"{ROOT_FOLDER}/Klasifikasi/Baseline_Report/Baseline_Confusion_Matrices"

# Running
X_train = df_train_cls.drop(columns="Kesegaran")
y_train = df_train_cls["Kesegaran"]

# Model yang digunakan
models = {
        "AdaBoost": AdaBoostClassifier(random_state=42),
        "DecisionTree": DecisionTreeClassifier(random_state=42),
        "KNN": KNeighborsClassifier(),
        "LogisticRegression": LogisticRegression(max_iter=2000, random_state=42),
        "RandomForest": RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1),
        "SVM": SVC(kernel="rbf", probability=True, random_state=42),
    }

baseline_summary = train_baseline_models(
    X_train=X_train,
    y_train=y_train,
    models=models,
    k_folds=10,
    random_state=42
)

baseline_summary
```

#### Feature Importance SHAP Method

Fungsi Helper SHAP

```python
def _get_estimator(model_or_pipeline: object) -> object:
    """
    Mengambil estimator terakhir dari sebuah Pipeline sklearn.

    Jika input berupa object Pipeline, fungsi akan mengembalikan
    estimator pada step terakhir. Jika bukan Pipeline, object
    akan dikembalikan apa adanya.

    Parameter
    ----------
    model_or_pipeline : object
        Estimator sklearn atau Pipeline sklearn.

    Returns
    -------
    object
        Estimator akhir yang akan digunakan untuk proses interpretasi
        atau analisis lanjutan.
    """
    if isinstance(model_or_pipeline, Pipeline):
        return model_or_pipeline.steps[-1][1]
    return model_or_pipeline


def _is_tree_model(estimator: object) -> bool:
    """
    Mendeteksi apakah estimator termasuk model berbasis tree.
    Deteksi dilakukan berdasarkan nama kelas estimator.

    Parameter
    ----------
    estimator : object
        Estimator sklearn atau model lain yang kompatibel.

    Returns
    -------
    bool
        True jika terdeteksi sebagai model tree-based,
        False jika bukan.
    """
    name = estimator.__class__.__name__.lower()
    tree_keywords = ["forest", "tree", "gradientboost", "extra", "xgb", "lgbm", "catboost"]
    return any(k in name for k in tree_keywords)


def _is_adaboost(estimator: object) -> bool:
    """
    Mendeteksi apakah estimator adalah AdaBoostClassifier dari sklearn.

    Parameter
    ----------
    estimator : object
        Estimator sklearn.

    Returns
    -------
    bool
        True jika estimator adalah AdaBoostClassifier,
        False jika bukan.
    """
    return estimator.__class__.__name__.lower() == "adaboostclassifier"


def _is_linear_model(estimator: object) -> bool:
    """
    Mendeteksi apakah estimator termasuk model linear.
    Deteksi dilakukan berdasarkan nama kelas.

    Parameter
    ----------
    estimator : object
        Estimator sklearn atau model lain.

    Returns
    -------
    bool
        True jika terdeteksi sebagai model linear,
        False jika bukan.
    """
    name = estimator.__class__.__name__.lower()
    linear_keywords = ["logisticregression", "linearsvc", "ridge", "lasso", "linearregression"]
    return any(k in name for k in linear_keywords)


def _to_mean_abs_importance(shap_values: object, feature_names: list) -> pd.DataFrame:
    """
    Mengonversi SHAP values menjadi nilai importance berbasis
    mean absolute SHAP per fitur.

    Fungsi ini mendukung berbagai format output SHAP:
    - shap.Explanation
    - list of arrays (multi-class)
    - numpy array 2D atau 3D

    Parameter
    ----------
    shap_values : array-like, list, atau shap.Explanation
        Output dari SHAP explainer.
    feature_names : Iterable[str]
        Daftar nama fitur yang sesuai dengan kolom input model.

    Returns
    -------
    pd.DataFrame
        DataFrame berisi dua kolom:
        - "Feature" : nama fitur
        - "MeanAbsSHAP" : rata-rata nilai absolut SHAP
        Diurutkan menurun berdasarkan nilai importance.
    """
    if hasattr(shap_values, "values"):
        shap_values = shap_values.values

    if isinstance(shap_values, list):
        abs_vals = np.mean([np.abs(v) for v in shap_values], axis=0)
        imp = abs_vals.mean(axis=0)
    else:
        sv = np.array(shap_values)
        if sv.ndim == 3:
            imp = np.mean(np.abs(sv), axis=(0, 2))
        else:
            imp = np.mean(np.abs(sv), axis=0)

    return (pd.DataFrame({"Feature": feature_names, "MeanAbsSHAP": imp})
              .sort_values("MeanAbsSHAP", ascending=False)
              .reset_index(drop=True))
```

Fungsi SHAP method dan Evaluasinya

```python
def shap_feature_importance_report(
    model_dict: dict,
    X: pd.DataFrame,
    y=None,
    max_samples: int = 1000,
    background_size: int = 200,
    random_state: int = 42
):
    """
    Menghasilkan laporan feature importance berbasis SHAP untuk beberapa model,
    serta menyimpan hasilnya dalam bentuk CSV dan gambar summary plot (bar).

    Fungsi ini melakukan sampling data untuk efisiensi, membangun background
    distribution untuk masker SHAP, lalu memilih metode explainer yang paling
    sesuai/aman berdasarkan tipe model:

    - TreeExplainer: digunakan untuk model tree-based (kecuali AdaBoost).
    - shap.Explainer (unified): fallback utama menggunakan masker Independent untuk menghindari
      error terkait DenseData / masking.
    - KernelExplainer: fallback terakhir jika dua metode di atas gagal.

    Parameter
    ----------
    model_dict : dict
        Dictionary berisi pasangan {nama_model: model atau Pipeline}.
        Jika value berupa Pipeline sklearn, estimator terakhir akan diambil.
    X : pd.DataFrame
        Data fitur yang akan digunakan untuk menghitung SHAP.
        Kolom pada X digunakan sebagai nama fitur (feature names).
    y : optional
        Disediakan untuk kompatibilitas pemanggilan (tidak digunakan pada implementasi saat ini).
    max_samples : int, default=1000
        Batas maksimum jumlah sampel dari X yang digunakan untuk menghitung SHAP.
        Jika len(X) > max_samples, data akan di-subsample secara acak tanpa pengembalian.
    background_size : int, default=200
        Ukuran background dataset untuk masker SHAP. Jika X lebih besar dari nilai ini,
        background akan di-subsample.
    random_state : int, default=42
        Seed untuk reproducibility pada proses sampling.

    Returns
    -------
    None
        Fungsi ini tidak mengembalikan nilai. Output disimpan ke file.

    Side Effects
    ------------
    Untuk setiap model pada `model_dict`, fungsi akan menyimpan:
    - CSV feature importance (mean absolute SHAP):
      f"{SHAP_DIR}/Features_Important_{model_name}.csv"
    - Gambar summary plot SHAP (bar):
      f"{SHAP_DIR}/Shap_Summary_{model_name}.jpg"
    Selain itu, fungsi akan mencetak status proses ke console.
    """

    rng = np.random.default_rng(random_state)

    # sampling X
    if len(X) > max_samples:
        sample_idx = rng.choice(X.index.to_numpy(), size=max_samples, replace=False)
        X_used = X.loc[sample_idx].copy()
    else:
        X_used = X.copy()

    feature_names = list(X_used.columns)

    # background untuk masker-
    bg = X_used
    if len(bg) > background_size:
        bg_idx = rng.choice(bg.index.to_numpy(), size=background_size, replace=False)
        bg = bg.loc[bg_idx].copy()

    # Ringkas background
    bg_mask = shap.sample(bg, nsamples=min(100, len(bg)), random_state=random_state)

    # masker independen
    masker = shap.maskers.Independent(bg_mask)

    for model_name, model_obj in model_dict.items():
        estimator = _get_estimator(model_obj)

        # fungsi prediksi untuk shap.Explainer/KernelExplainer
        pred_fn = estimator.predict_proba if hasattr(estimator, "predict_proba") else estimator.predict

        shap_values = None
        explain_mode = None

        # TreeExplainer untuk tree model (kecuali AdaBoost)
        try:
            if _is_tree_model(estimator) and (not _is_adaboost(estimator)):
                explain_mode = "TreeExplainer"
                explainer = shap.TreeExplainer(estimator)
                shap_values = explainer(X_used)
        except Exception:
            shap_values = None

        # Unified Explainer
        if shap_values is None:
            try:
                explain_mode = "shap.Explainer"
                explainer = shap.Explainer(pred_fn, masker)
                shap_values = explainer(X_used)
            except Exception:
                shap_values = None

        # KernelExplainer fallback
        if shap_values is None:
            try:
                explain_mode = "KernelExplainer"
                bg_kernel = bg_mask
                explainer = shap.KernelExplainer(pred_fn, bg_kernel)
                shap_values = explainer.shap_values(X_used)
            except Exception as e:
                print(f"[SKIP] SHAP gagal untuk {model_name}. Alasan: {e}")
                continue

        # Save CSV feature importance
        df_imp = _to_mean_abs_importance(shap_values, feature_names)
        csv_path = f"{SHAP_DIR}/Features_Important_{model_name}.csv"
        df_imp.to_csv(csv_path, index=False)

        #  Save summary plot (bar)
        plt.figure(figsize=(10, 6))
        values_for_plot = shap_values.values if hasattr(shap_values, "values") else shap_values

        shap.summary_plot(
            values_for_plot,
            X_used,
            feature_names=feature_names,
            show=False,
            plot_type="bar"
        )
        plt.title(f"SHAP Feature Importance - {model_name} ({explain_mode})")
        jpg_path = f"{SHAP_DIR}/Shap_Summary_{model_name}.jpg"
        plt.tight_layout()
        plt.savefig(jpg_path, dpi=200)
        plt.close()

        print(f"[OK] {model_name}: saved -> {jpg_path} and {csv_path}")
```

#### Re-Training dan Evaluation

Fungsi Load Top-N Features dari CSV SHAP Importance

```python
def load_top_features_from_shap_csv(shap_csv_path: str, n_features: int) -> list:
    """
    Membaca file CSV hasil SHAP feature importance dan
    mengambil sejumlah fitur teratas berdasarkan urutan.

    File CSV diasumsikan memiliki minimal dua kolom:
    - "Feature"
    - "MeanAbsSHAP"

    Parameter
    ----------
    shap_csv_path : str
        Path menuju file CSV hasil SHAP feature importance
        (misalnya: Features_Important_[Nama-Model].csv).
    n_features : int
        Jumlah fitur teratas (top-n) yang ingin diambil.

    Returns
    -------
    list
        List berisi nama-nama fitur teratas berdasarkan
        urutan pada file CSV yang telah diurutkan.
    """

    df = pd.read_csv(shap_csv_path)
    if "Feature" not in df.columns:
        raise ValueError(f"Kolom 'Feature' tidak ditemukan pada {shap_csv_path}. Kolom ada: {list(df.columns)}")
    top_feats = df["Feature"].head(n_features).tolist()
    return top_feats
```

Fungsi Evaluasi Cross Validation

```python
def evaluate_model_cv_and_save(
    model_name: str,
    estimator_or_pipeline,
    X: pd.DataFrame,
    y: pd.Series,
    n_features: int,
    k_folds: int = 10,
    random_state: int = 42,
    save_dir_cm: str = None
):
    """
    Mengevaluasi model klasifikasi menggunakan Stratified K-Fold
    cross-validation dan menyimpan hasil evaluasi ke file.

    Fungsi ini akan:
    1) Melakukan prediksi cross-validation dengan `cross_val_predict`.
    2) Membuat dan menyimpan confusion matrix dalam format gambar (.jpg).
    3) Menyimpan classification report lengkap dalam format CSV.
    4) Mengembalikan ringkasan metrik utama dalam bentuk dictionary.

    Parameter
    ----------
    model_name : str
        Nama model yang digunakan untuk penamaan file output.
    estimator_or_pipeline : object
        Estimator sklearn atau Pipeline yang akan dievaluasi.
    X : pd.DataFrame
        Data fitur untuk evaluasi.
    y : pd.Series
        Target/label untuk evaluasi.
    n_features : int
        Jumlah fitur yang digunakan (untuk keperluan penamaan file).
    k_folds : int, default=10
        Jumlah fold untuk Stratified K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada StratifiedKFold (shuffle=True).
    save_dir_cm : str, optional
        Direktori penyimpanan file confusion matrix dan metrics CSV.
        Jika None, akan menggunakan variabel global `RETRAIN_CM_DIR`.

    Returns
    -------
    dict
        Dictionary ringkasan evaluasi berisi:
        - "Model"
        - "Num_Features"
        - "Accuracy"
        - "Precision_weighted"
        - "Recall_weighted"
        - "F1_weighted"
        - "ConfusionMatrixPath"
        - "MetricsCSVPath"

    Side Effects
    ------------
    - Menyimpan file confusion matrix:
      Confusion_Matrix_[Model]_[Jumlah-Fitur]-Features.jpg
    - Menyimpan file classification report:
      Metric_Evaluation_[Model]_[Jumlah-Fitur]-Features.csv
    - Mencetak status penyimpanan ke console.
    """

    if save_dir_cm is None:
        save_dir_cm = RETRAIN_CM_DIR

    cv = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=random_state)

    # cross_val_predict untuk confusion matrix dari prediksi CV
    y_pred = cross_val_predict(estimator_or_pipeline, X, y, cv=cv, n_jobs=-1)

    # Confusion Matrix
    cm = confusion_matrix(y, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name} ({n_features} Features)")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")

    cm_path = f"{save_dir_cm}/Confusion_Matrix_{model_name}_{n_features}-Features.jpg"
    plt.tight_layout()
    plt.savefig(cm_path, dpi=200)
    plt.close()

    # Metrics CSV (classification_report detail)
    metric_df = pd.DataFrame(classification_report(y, y_pred, output_dict=True)).transpose()
    metric_csv_path = f"{save_dir_cm}/Metric_Evaluation_{model_name}_{n_features}-Features.csv"
    metric_df.to_csv(metric_csv_path)

    # Return ringkasan cepat (opsional)
    summary = {
        "Model": model_name,
        "Num_Features": n_features,
        "Accuracy": float(accuracy_score(y, y_pred)),
        "Precision_weighted": float(precision_score(y, y_pred, average="weighted", zero_division=0)),
        "Recall_weighted": float(recall_score(y, y_pred, average="weighted", zero_division=0)),
        "F1_weighted": float(f1_score(y, y_pred, average="weighted", zero_division=0)),
        "ConfusionMatrixPath": cm_path,
        "MetricsCSVPath": metric_csv_path
    }

    print(f"[OK] CV Eval saved -> {cm_path} & {metric_csv_path}")
    return summary
```

Fungsi Save ONNX

```python
def fit_full_and_export_onnx(
    model_name: str,
    estimator_or_pipeline,
    X: pd.DataFrame,
    y: pd.Series,
    n_features: int,
    save_dir_onnx: str = None
):
    """
    Melatih model pada seluruh data training (X, y) kemudian
    mengekspor model ke format ONNX.

    File output akan dinamai:
    Model_[Nama-Model]_[Jumlah-Fitur]-Features.onnx

    Parameter
    ----------
    model_name : str
        Nama model yang digunakan untuk penamaan file output.
    estimator_or_pipeline : object
        Estimator sklearn atau Pipeline yang akan dilatih dan diekspor.
    X : pd.DataFrame
        Data fitur training berukuran (n_samples, n_features).
        Diasumsikan seluruh fitur numerik (float) untuk kompatibilitas ONNX.
    y : pd.Series
        Target/label training berukuran (n_samples,).
    n_features : int
        Jumlah fitur yang digunakan (untuk keperluan penamaan file).
    save_dir_onnx : str, optional
        Direktori penyimpanan file ONNX.
        Jika None, akan menggunakan variabel global `RETRAIN_ONNX_DIR`.

    Returns
    -------
    str
        Path lengkap file ONNX yang berhasil disimpan.

    Side Effects
    ------------
    - Melatih ulang model menggunakan seluruh data (tanpa cross-validation).
    - Menyimpan file ONNX ke direktori yang ditentukan.
    - Mencetak status penyimpanan ke console.
    """

    if save_dir_onnx is None:
        save_dir_onnx = RETRAIN_ONNX_DIR

    estimator_or_pipeline.fit(X, y)

    initial_type = [("float_input", FloatTensorType([None, X.shape[1]]))]
    onnx_model = convert_sklearn(estimator_or_pipeline, initial_types=initial_type, target_opset=12)

    onnx_path = f"{save_dir_onnx}/Model_{model_name}_{n_features}-Features.onnx"
    with open(onnx_path, "wb") as f:
        f.write(onnx_model.SerializeToString())

    print(f"[OK] ONNX saved -> {onnx_path}")
    return onnx_path
```

Fungsi Re-Training Model

```python
def retrain_models_with_shap_features(
    models: dict,
    X_train: pd.DataFrame,
    y_train: pd.Series,
    shap_dir: str,
    feature_range=range(3, 17),
    k_folds: int = 10,
    random_state: int = 42
) -> pd.DataFrame:
    """
    Melakukan retraining model menggunakan subset fitur terbaik
    berdasarkan hasil SHAP feature importance.

    Untuk setiap model dalam variabel`models`, fungsi ini akan:
    1) Membaca file SHAP CSV: Features_Important_[Nama-Model].csv.
    2) Mengambil top-n fitur sesuai `feature_range`.
    3) Melakukan evaluasi Stratified K-Fold CV untuk setiap jumlah fitur.
    4) Menyimpan confusion matrix dan metrics CSV.
    5) Melatih ulang model pada seluruh data subset fitur.
    6) Mengekspor model hasil retraining ke format ONNX.
    7) Mengumpulkan seluruh ringkasan hasil ke dalam DataFrame.

    Parameter
    ----------
    models : dict
        Dictionary berisi pasangan {nama_model: estimator/pipeline}.
        Estimator harus kompatibel dengan sklearn dan dapat di-clone.
        Model akan di-fit ulang untuk setiap subset fitur.
    X_train : pd.DataFrame
        Data fitur training penuh.
    y_train : pd.Series
        Target/label training.
    shap_dir : str
        Direktori yang berisi file SHAP CSV dengan format:
        Features_Important_[Nama-Model].csv
    feature_range : iterable, default=range(3, 17)
        Rentang jumlah fitur yang akan diuji.
    k_folds : int, default=10
        Jumlah fold untuk Stratified K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada proses cross-validation.

    Returns
    -------
    pd.DataFrame
        DataFrame berisi ringkasan hasil evaluasi untuk setiap
        kombinasi model dan jumlah fitur, termasuk path file
        confusion matrix, metrics CSV, dan ONNX.

    Side Effects
    ------------
    - Membaca file SHAP CSV dari `shap_dir`.
    - Menyimpan confusion matrix dan metrics CSV ke `RETRAIN_CM_DIR`.
    - Menyimpan model ONNX ke `RETRAIN_ONNX_DIR`.
    - Mencetak status proses ke console.
    """

    results = []

    for model_name, model_obj in models.items():
        shap_csv_path = f"{shap_dir}/Features_Important_{model_name}.csv"
        if not os.path.exists(shap_csv_path):
            print(f"[SKIP] SHAP CSV tidak ditemukan untuk {model_name}: {shap_csv_path}")
            continue

        # Clone estimator/pipeline
        from sklearn.base import clone
        base_model = clone(model_obj)

        for n_feats in feature_range:
            # Ambil top-n features dari SHAP CSV
            top_feats = load_top_features_from_shap_csv(shap_csv_path, n_feats)

            # Subset data
            X_sub = X_train[top_feats].copy()

            # Clone untuk menjaga keaslian model
            model_to_use = clone(base_model)

            # Evaluasi CV dan simpan CM & Metrics
            summary = evaluate_model_cv_and_save(
                model_name=model_name,
                estimator_or_pipeline=model_to_use,
                X=X_sub,
                y=y_train,
                n_features=n_feats,
                k_folds=k_folds,
                random_state=random_state,
                save_dir_cm=RETRAIN_CM_DIR
            )

            # Fit full dan export ONNX
            onnx_path = fit_full_and_export_onnx(
                model_name=model_name,
                estimator_or_pipeline=model_to_use,
                X=X_sub,
                y=y_train,
                n_features=n_feats,
                save_dir_onnx=RETRAIN_ONNX_DIR
            )

            summary["ONNXPath"] = onnx_path
            results.append(summary)

    return pd.DataFrame(results)
```

```python
# Target Folder Penyimpanan hasil
RETRAIN_ROOT = f"{ROOT_FOLDER}/Klasifikasi/Retraining_Report"
RETRAIN_CM_DIR = f"{RETRAIN_ROOT}/Confusion_Matrices"
RETRAIN_ONNX_DIR = f"{RETRAIN_ROOT}/ONNX_Models"

# Running
Final_X_train = df_train_cls
Final_y_train = df_train_cls["Kesegaran"]

# Folder SHAP Report
SHAP_DIR = f"{ROOT_FOLDER}/Klasifikasi/SHAP_Report"

retrain_results_df = retrain_models_with_shap_features(
    models=models,
    X_train=Final_X_train,
    y_train=Final_y_train,
    shap_dir=SHAP_DIR,
    feature_range=range(3, 17),
    k_folds=10,
    random_state=42
)

# Simpan ringkasan semua hasil retraining
summary_path = f"{RETRAIN_ROOT}/Retraining_Summary_3-16_Features.csv"
retrain_results_df.to_csv(summary_path, index=False)

print("\n[OK] Retraining summary saved:", summary_path)
retrain_results_df
```

```python
def pre_processing_regresi(
    df: pd.DataFrame,
    columns_to_drop: Iterable[str] = (
        "Kesegaran",
        "Hour",
        "Source_Image",
        "Augment_ID",
    )
) -> pd.DataFrame:
    """
    Melakukan preprocessing awal untuk data regresi dengan
    menghapus kolom-kolom tertentu jika tersedia pada DataFrame.

    Parameter
    ----------
    df : pd.DataFrame
        DataFrame yang akan diproses.
    columns_to_drop : Iterable[str], optional
        Daftar nama kolom yang ingin dihapus dari DataFrame.
        Kolom hanya akan dihapus jika terdapat dalam DataFrame.

    Returns
    -------
    pd.DataFrame
        DataFrame hasil preprocessing dengan kolom yang ditentukan
        telah dihapus. DataFrame asli tidak dimodifikasi
        karena dilakukan proses copy terlebih dahulu.
    """

    processed_df = df.copy()

    # Hanya drop kolom yang tersedia
    existing_cols = [col for col in columns_to_drop if col in processed_df.columns]

    if existing_cols:
        processed_df = processed_df.drop(columns=existing_cols)

    return processed_df
```

Fungsi Training dan Evaluasi Baseline Model

```python
def train_baseline_models_regression(
    X_train: pd.DataFrame,
    y_train: pd.Series,
    models: dict,
    k_folds: int = 10,
    random_state: int = 42,
    export_pkl: bool = True
):
    """
    Melatih dan mengevaluasi beberapa model baseline regresi menggunakan
    K-Fold cross-validation, lalu melatih model final pada seluruh data
    training dan mengekspor model ke format ONNX dan PKL.

    Untuk setiap model pada variabel `models`, fungsi ini akan:
    1) Membuat Pipeline([("model", estimator)]).
    2) Melakukan prediksi cross-validation dengan `cross_val_predict`.
    3) Menghitung metrik regresi: MAE, RMSE, dan R2 pada prediksi out-of-fold.
    4) Menyimpan metrik per model ke file CSV.
    5) Melatih model final pada seluruh data training.
    6) Mengekspor model final ke ONNX.
    7) Menyimpan model final dalam format PKL menggunakan joblib.
    8) Menyimpan ringkasan seluruh model ke CSV (diurutkan berdasarkan RMSE).

    Parameter
    ----------
    X_train : pd.DataFrame
        Fitur data training berukuran (n_samples, n_features).
        Diasumsikan fitur numerik dan kompatibel dengan ekspor ONNX
        menggunakan FloatTensorType.
    y_train : pd.Series
        Target regresi berukuran (n_samples,).
    models : dict
        Dictionary berisi pasangan {nama_model: estimator_sklearn}.
    k_folds : int, default=10
        Jumlah fold untuk K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada KFold (shuffle=True).
    export_pkl : bool, default=True
        Jika True, model final juga disimpan dalam format PKL untuk digunakan di Python.

    Returns
    -------
    pd.DataFrame
        DataFrame ringkasan performa setiap model dengan kolom:
        ["Model", "MAE", "RMSE", "R2"], diurutkan menaik berdasarkan RMSE
        (semakin kecil semakin baik). Ringkasan juga disimpan ke file CSV.

    Side Effects
    ------------
    - Menyimpan metrik per model:
      f"{EVAL_DIR}/Metric_Evaluation_{model_name}.csv"
    - Menyimpan model ONNX per model:
      f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.onnx"
    - (Opsional) Menyimpan model PKL per model:
      f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.pkl"
    - Menyimpan ringkasan seluruh model:
      f"{REPORT_DIR}/Baseline_Summary_Report.csv"
    - Mencetak status proses ke console.
    """

    # CV untuk regresi
    cv = KFold(n_splits=k_folds, shuffle=True, random_state=random_state)

    summary_rows = []

    for model_name, estimator in models.items():

        # Pipeline sederhana
        pipe = Pipeline([("model", estimator)])


        # Cross Validation Prediction
        y_pred_cv = cross_val_predict(pipe, X_train, y_train, cv=cv, n_jobs=-1)

        # Metrics
        mae = mean_absolute_error(y_train, y_pred_cv)
        rmse = np.sqrt(mean_squared_error(y_train, y_pred_cv))
        r2 = r2_score(y_train, y_pred_cv)

        summary_rows.append({
            "Model": model_name,
            "MAE": float(mae),
            "RMSE": float(rmse),
            "R2": float(r2)
        })

        # Save per-model metric CSV
        metric_df = pd.DataFrame([{
            "Model": model_name,
            "CV_Folds": k_folds,
            "MAE": float(mae),
            "RMSE": float(rmse),
            "R2": float(r2)
        }])

        metric_csv_path = f"{EVAL_DIR}/Metric_Evaluation_{model_name}.csv"
        metric_df.to_csv(metric_csv_path, index=False)

        # Fit Final Model
        pipe.fit(X_train, y_train)

        # Save ONNX
        initial_type = [("float_input", FloatTensorType([None, X_train.shape[1]]))]
        onnx_model = convert_sklearn(pipe, initial_types=initial_type, target_opset=12)

        onnx_path = f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.onnx"
        with open(onnx_path, "wb") as f:
            f.write(onnx_model.SerializeToString())

        # Save PKL
        if export_pkl:
            pkl_path = f"{BASELINE_MODEL_DIR}/Baseline_Model_{model_name}.pkl"
            joblib.dump(pipe, pkl_path)

        print(f"[DONE] {model_name} selesai disimpan.")

    # Summary CSV semua model
    summary_df = pd.DataFrame(summary_rows).sort_values(by="RMSE", ascending=True)  # kecil lebih baik
    summary_df.to_csv(f"{REPORT_DIR}/Baseline_Summary_Report.csv", index=False)

    return summary_df
```

#### Feature Importance SHAP Method

Fungsi Helper SHAP

```python
def _get_estimator(model_or_pipeline):
    """
    Mengambil estimator terakhir dari sebuah Pipeline sklearn.

    Jika input berupa object Pipeline, fungsi akan mengembalikan
    estimator pada step terakhir. Jika bukan Pipeline, object
    akan dikembalikan apa adanya.

    Parameter
    ----------
    model_or_pipeline : object
        Estimator sklearn atau Pipeline sklearn.

    Returns
    -------
    object
        Estimator akhir yang siap digunakan untuk training,
        evaluasi, atau interpretasi (misalnya SHAP).
    """

    if isinstance(model_or_pipeline, Pipeline):
        return model_or_pipeline.steps[-1][1]
    return model_or_pipeline


def _is_tree_model(estimator) -> bool:
    """
    Mendeteksi apakah estimator termasuk model berbasis tree.

    Deteksi dilakukan berdasarkan nama kelas estimator dengan
    mencari keyword.

    Parameter
    ----------
    estimator : object
        Estimator sklearn atau model lain.

    Returns
    -------
    bool
        True jika terdeteksi sebagai model tree-based,
        False jika bukan.
    """

    name = estimator.__class__.__name__.lower()
    tree_keywords = ["forest", "tree", "gradientboost", "extra", "xgb", "lgbm", "catboost"]
    return any(k in name for k in tree_keywords)


def _is_linear_model(estimator) -> bool:
    """
    Mendeteksi apakah estimator termasuk model linear.

    Deteksi dilakukan berdasarkan nama kelas estimator dengan
    mencari keyword.

    Parameter
    ----------
    estimator : object
        Estimator sklearn atau model lain.

    Returns
    -------
    bool
        True jika terdeteksi sebagai model linear,
        False jika bukan.
    """

    name = estimator.__class__.__name__.lower()
    linear_keywords = ["linearregression", "ridge", "lasso", "elasticnet", "sgdregressor"]
    return any(k in name for k in linear_keywords)


def _to_mean_abs_importance(shap_values, feature_names):
    """
    Mengonversi SHAP values menjadi nilai feature importance
    berbasis mean absolute SHAP per fitur.

    Fungsi ini mendukung berbagai format output SHAP:
    - shap.Explanation
    - list of arrays
    - numpy array 2D (n_samples, n_features)
    - numpy array 3D (misalnya multi-output)

    Parameter
    ----------
    shap_values : array-like, list, atau shap.Explanation
        Output dari SHAP explainer.
    feature_names : Iterable[str]
        Daftar nama fitur yang sesuai dengan kolom input model.

    Returns
    -------
    pd.DataFrame
        DataFrame berisi dua kolom:
        - "Feature" : nama fitur
        - "MeanAbsSHAP" : rata-rata nilai absolut SHAP
        Diurutkan menurun berdasarkan importance.
    """

    # shap.Explanation
    if hasattr(shap_values, "values"):
        shap_values = shap_values.values

    # Handling Eksplainer Regresi ndim=3
    if isinstance(shap_values, list):
        abs_vals = np.mean([np.abs(v) for v in shap_values], axis=0)
        imp = abs_vals.mean(axis=0)
    else:
        sv = np.array(shap_values)
        if sv.ndim == 3:
            imp = np.mean(np.abs(sv), axis=(0, 2))
        else:
            imp = np.mean(np.abs(sv), axis=0)

    return (pd.DataFrame({"Feature": feature_names, "MeanAbsSHAP": imp})
              .sort_values("MeanAbsSHAP", ascending=False)
              .reset_index(drop=True))
```

Fungsi SHAP method dan Evaluasinya

```python
def shap_feature_importance_report(
    model_dict: dict,
    X: pd.DataFrame,
    y=None,
    max_samples: int = 1000,
    background_size: int = 200,
    random_state: int = 42
):
    """
    Menghasilkan laporan feature importance berbasis SHAP untuk model regresi,
    serta menyimpan hasilnya dalam bentuk CSV dan gambar summary plot (bar).

    Fungsi ini melakukan sampling data untuk efisiensi, membangun background
    distribution untuk masker SHAP (tabular), lalu memilih metode explainer
    yang paling sesuai berdasarkan tipe model:

    - TreeExplainer: digunakan untuk model tree-based.
    - shap.Explainer (unified): fallback utama menggunakan masker Independent.
    - KernelExplainer: fallback terakhir.

    Parameter
    ----------
    model_dict : dict
        Dictionary berisi pasangan {nama_model: model atau Pipeline}.
        Jika value berupa Pipeline sklearn, estimator terakhir akan diambil.
    X : pd.DataFrame
        Data fitur yang akan digunakan untuk menghitung SHAP.
        Kolom pada X digunakan sebagai nama fitur (feature names).
    y : optional
        Disediakan untuk kompatibilitas pemanggilan.
    max_samples : int, default=1000
        Batas maksimum jumlah sampel dari X yang digunakan untuk menghitung SHAP.
        Jika len(X) > max_samples, data akan di-subsample secara acak tanpa pengembalian.
    background_size : int, default=200
        Ukuran background dataset untuk masker SHAP. Jika X lebih besar dari nilai ini,
        background akan di-subsample.
    random_state : int, default=42
        Seed untuk reproducibility pada proses sampling.

    Returns
    -------
    None

    Side Effects
    ------------
    Untuk setiap model pada `model_dict`, fungsi akan menyimpan:
    - CSV feature importance (mean absolute SHAP):
      f"{SHAP_DIR}/Features_Important_{model_name}.csv"
    - Gambar summary plot SHAP (bar):
      f"{SHAP_DIR}/Shap_Summary_{model_name}.jpg"
    Selain itu, fungsi akan mencetak status proses ke console.
    """

    rng = np.random.default_rng(random_state)

    # Sampling X
    if len(X) > max_samples:
        sample_idx = rng.choice(X.index.to_numpy(), size=max_samples, replace=False)
        X_used = X.loc[sample_idx].copy()
    else:
        X_used = X.copy()

    feature_names = list(X_used.columns)

    # Background untuk masker
    bg = X_used
    if len(bg) > background_size:
        bg_idx = rng.choice(bg.index.to_numpy(), size=background_size, replace=False)
        bg = bg.loc[bg_idx].copy()

    # Ringkas background
    bg_mask = shap.sample(bg, nsamples=min(100, len(bg)), random_state=random_state)

    # Masker independen (tabular)
    masker = shap.maskers.Independent(bg_mask)

    for model_name, model_obj in model_dict.items():
        estimator = _get_estimator(model_obj)

        # Predict Regresi
        pred_fn = estimator.predict

        shap_values = None
        explain_mode = None

        # TreeExplainer untuk tree model
        try:
            if _is_tree_model(estimator):
                explain_mode = "TreeExplainer"
                explainer = shap.TreeExplainer(estimator)
                shap_values = explainer(X_used)
        except Exception:
            shap_values = None

        # Unified Explainer fallback
        if shap_values is None:
            try:
                explain_mode = "shap.Explainer"
                explainer = shap.Explainer(pred_fn, masker)
                shap_values = explainer(X_used)
            except Exception:
                shap_values = None

        # KernelExplainer fallback
        if shap_values is None:
            try:
                explain_mode = "KernelExplainer"
                bg_kernel = bg_mask
                explainer = shap.KernelExplainer(pred_fn, bg_kernel)
                shap_values = explainer.shap_values(X_used)
            except Exception as e:
                print(f"[SKIP] SHAP gagal untuk {model_name}. Alasan: {e}")
                continue

        # Save CSV feature importance
        df_imp = _to_mean_abs_importance(shap_values, feature_names)
        csv_path = f"{SHAP_DIR}/Features_Important_{model_name}.csv"
        df_imp.to_csv(csv_path, index=False)

        # Save summary plot (bar)
        plt.figure(figsize=(10, 6))
        values_for_plot = shap_values.values if hasattr(shap_values, "values") else shap_values

        shap.summary_plot(
            values_for_plot,
            X_used,
            feature_names=feature_names,
            show=False,
            plot_type="bar"
        )
        plt.title(f"SHAP Feature Importance (Regression) - {model_name} ({explain_mode})")
        jpg_path = f"{SHAP_DIR}/Shap_Summary_{model_name}.jpg"
        plt.tight_layout()
        plt.savefig(jpg_path, dpi=200)
        plt.close()

        print(f"[OK] {model_name}: saved -> {jpg_path} and {csv_path}")
```

#### Re-Training dan Evaluation

Fungsi Load Top-N Features dari CSV SHAP Importance

```python
def load_top_features_from_shap_csv(shap_csv_path: str, n_features: int) -> list:
    """
    Membaca file CSV hasil SHAP feature importance dan
    mengambil sejumlah fitur teratas (top-n).

    File CSV diasumsikan memiliki minimal dua kolom:
    - "Feature"
    - "MeanAbsSHAP"

    Parameter
    ----------
    shap_csv_path : str
        Path menuju file CSV hasil SHAP feature importance
    n_features : int
        Jumlah fitur teratas (top-n) yang ingin diambil.

    Returns
    -------
    list
        List berisi nama-nama fitur teratas berdasarkan
        urutan pada file CSV (diasumsikan sudah diurutkan
        menurun berdasarkan MeanAbsSHAP).
    """

    df = pd.read_csv(shap_csv_path)
    if "Feature" not in df.columns:
        raise ValueError(f"Kolom 'Feature' tidak ditemukan pada {shap_csv_path}. Kolom ada: {list(df.columns)}")
    top_feats = df["Feature"].head(n_features).tolist()
    return top_feats
```

Fungsi Evaluasi Cross Validation

```python
def evaluate_model_cv_and_save_regression(
    model_name: str,
    estimator_or_pipeline,
    X: pd.DataFrame,
    y: pd.Series,
    n_features: int,
    k_folds: int = 10,
    random_state: int = 42,
    save_dir_eval: str = None
):
    """
    Mengevaluasi model regresi menggunakan K-Fold cross-validation
    dan menyimpan hasil evaluasi ke file.

    Fungsi ini akan:
    1) Melakukan prediksi cross-validation dengan `cross_val_predict`.
    2) Menghitung metrik regresi: MAE, RMSE, dan R2.
    3) Menyimpan metrik dalam format CSV.
    4) Membuat dan menyimpan plot Actual vs Predicted.

    Parameter
    ----------
    model_name : str
        Nama model yang digunakan untuk penamaan file output.
    estimator_or_pipeline : object
        Estimator sklearn atau Pipeline yang akan dievaluasi.
    X : pd.DataFrame
        Data fitur untuk evaluasi.
    y : pd.Series
        Target regresi.
    n_features : int
        Jumlah fitur yang digunakan (untuk keperluan penamaan file).
    k_folds : int, default=10
        Jumlah fold untuk K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada KFold (shuffle=True).
    save_dir_eval : str, optional
        Direktori penyimpanan file evaluasi.
        Jika None, akan menggunakan variabel global `RETRAIN_EVAL_DIR`.

    Returns
    -------
    dict
        Dictionary ringkasan evaluasi berisi:
        - "Model"
        - "Num_Features"
        - "MAE"
        - "RMSE"
        - "R2"
        - "MetricsCSVPath"
        - "PredictionPlotPath"

    Side Effects
    ------------
    - Menyimpan file metrik:
      Metric_Evaluation_[Model]_[Jumlah-Fitur]-Features.csv
    - Menyimpan file plot:
      Prediction_Plot_[Model]_[Jumlah-Fitur]-Features.jpg
    - Mencetak status penyimpanan ke console.
    """

    if save_dir_eval is None:
        save_dir_eval = RETRAIN_EVAL_DIR

    cv = KFold(n_splits=k_folds, shuffle=True, random_state=random_state)

    # cross_val_predict untuk prediksi CV
    y_pred = cross_val_predict(estimator_or_pipeline, X, y, cv=cv, n_jobs=-1)

    # Metrics
    mae = float(mean_absolute_error(y, y_pred))
    rmse = float(np.sqrt(mean_squared_error(y, y_pred)))
    r2 = float(r2_score(y, y_pred))

    # Save metrics CSV per model-per-feature
    metric_df = pd.DataFrame([{
        "Model": model_name,
        "Num_Features": n_features,
        "CV_Folds": k_folds,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2
    }])

    metric_csv_path = f"{save_dir_eval}/Metric_Evaluation_{model_name}_{n_features}-Features.csv"
    metric_df.to_csv(metric_csv_path, index=False)

    # Save plot (Actual vs Predicted)
    plt.figure(figsize=(6, 6))
    plt.scatter(y, y_pred, alpha=0.6)
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"Actual vs Predicted - {model_name} ({n_features} Features)")
    plt.tight_layout()

    plot_path = f"{save_dir_eval}/Prediction_Plot_{model_name}_{n_features}-Features.jpg"
    plt.savefig(plot_path, dpi=200)
    plt.close()

    summary = {
        "Model": model_name,
        "Num_Features": n_features,
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2,
        "MetricsCSVPath": metric_csv_path,
        "PredictionPlotPath": plot_path
    }

    print(f"[OK] CV Eval saved -> {metric_csv_path} & {plot_path}")
    return summary
```

Fungsi Save ONNX

```python
def fit_full_and_export_onnx_regression(
    model_name: str,
    estimator_or_pipeline,
    X: pd.DataFrame,
    y: pd.Series,
    n_features: int,
    save_dir_onnx: str = None
):
    """
    Melatih model regresi pada seluruh data training (X, y) kemudian
    mengekspor model ke format ONNX.

    File output akan dinamai:
    Model_[Nama-Model]_[Jumlah-Fitur]-Features.onnx

    Parameter
    ----------
    model_name : str
        Nama model yang digunakan untuk penamaan file output.
    estimator_or_pipeline : object
        Estimator sklearn atau Pipeline yang akan dilatih dan diekspor.
    X : pd.DataFrame
        Data fitur training berukuran (n_samples, n_features).
        Diasumsikan fitur numerik dan kompatibel untuk input ONNX bertipe float.
    y : pd.Series
        Target regresi training berukuran (n_samples,).
    n_features : int
        Jumlah fitur yang digunakan (untuk keperluan penamaan file).
    save_dir_onnx : str, optional
        Direktori penyimpanan file ONNX.
        Jika None, akan menggunakan variabel global `RETRAIN_ONNX_DIR`.

    Returns
    -------
    str
        Path lengkap file ONNX yang berhasil disimpan.

    Side Effects
    ------------
    - Melatih ulang model menggunakan seluruh data.
    - Menyimpan file ONNX ke direktori yang ditentukan.
    - Mencetak status penyimpanan ke console.
    """

    if save_dir_onnx is None:
        save_dir_onnx = RETRAIN_ONNX_DIR

    estimator_or_pipeline.fit(X, y)

    initial_type = [("float_input", FloatTensorType([None, X.shape[1]]))]
    onnx_model = convert_sklearn(estimator_or_pipeline, initial_types=initial_type, target_opset=12)

    onnx_path = f"{save_dir_onnx}/Model_{model_name}_{n_features}-Features.onnx"
    with open(onnx_path, "wb") as f:
        f.write(onnx_model.SerializeToString())

    print(f"[OK] ONNX saved -> {onnx_path}")
    return onnx_path
```

Fungsi Retraining Model

```python
def retrain_models_with_shap_features_regression(
    models: dict,
    X_train: pd.DataFrame,
    y_train: pd.Series,
    shap_dir: str,
    feature_range=range(3, 17),
    k_folds: int = 10,
    random_state: int = 42
) -> pd.DataFrame:
    """
    Melakukan retraining model regresi menggunakan subset fitur terbaik
    berdasarkan hasil SHAP feature importance, lalu mengevaluasi dengan
    K-Fold CV dan mengekspor model ke ONNX untuk setiap jumlah fitur.

    Untuk setiap model pada variabel `models`, fungsi ini akan:
    1) Membaca file SHAP CSV: Features_Important_[Nama-Model].csv dari `shap_dir`.
    2) Mengambil top-n fitur untuk setiap n pada `feature_range`.
    3) Membuat subset data X_train berdasarkan fitur terpilih.
    4) Mengevaluasi model dengan K-Fold CV (MAE, RMSE, R2), menyimpan:
       - CSV metrik
       - Plot Actual vs Predicted
    5) Melatih ulang model pada seluruh data subset fitur dan mengekspor ONNX.
    6) Mengumpulkan ringkasan hasil ke dalam DataFrame.

    Parameter
    ----------
    models : dict
        Dictionary berisi pasangan {nama_model: estimator/pipeline}.
        Estimator harus kompatibel dengan sklearn dan dapat di-clone.
        Model akan di-fit ulang untuk setiap subset fitur.
    X_train : pd.DataFrame
        Data fitur training penuh (sebelum seleksi SHAP).
    y_train : pd.Series
        Target regresi training.
    shap_dir : str
        Direktori yang berisi file SHAP CSV dengan format:
        Features_Important_[Nama-Model].csv
    feature_range : iterable, default=range(3, 17)
        Rentang jumlah fitur yang akan diuji.
    k_folds : int, default=10
        Jumlah fold untuk K-Fold cross-validation.
    random_state : int, default=42
        Seed untuk reproducibility pada proses cross-validation.

    Returns
    -------
    pd.DataFrame
        DataFrame berisi ringkasan hasil evaluasi untuk setiap kombinasi
        (model, jumlah fitur), termasuk metrik (MAE/RMSE/R2) dan path file
        output (metrics CSV, prediction plot, dan ONNX).

    Side Effects
    ------------
    - Membaca file SHAP CSV dari `shap_dir`.
    - Menyimpan file evaluasi ke `RETRAIN_EVAL_DIR` (CSV + plot).
    - Menyimpan model ONNX ke `RETRAIN_ONNX_DIR`.
    - Mencetak status proses ke console.
    """

    results = []

    for model_name, model_obj in models.items():
        shap_csv_path = f"{shap_dir}/Features_Important_{model_name}.csv"
        if not os.path.exists(shap_csv_path):
            print(f"[SKIP] SHAP CSV tidak ditemukan untuk {model_name}: {shap_csv_path}")
            continue

        base_model = clone(model_obj)

        for n_feats in feature_range:
            # Ambil top-n features dari SHAP
            top_feats = load_top_features_from_shap_csv(shap_csv_path, n_feats)

            # Subset data (fitur saja)
            X_sub = X_train[top_feats].copy()

            # Clone model untuk tiap N fitur
            model_to_use = clone(base_model)

            # (i) Evaluasi CV + simpan metric & plot
            summary = evaluate_model_cv_and_save_regression(
                model_name=model_name,
                estimator_or_pipeline=model_to_use,
                X=X_sub,
                y=y_train,
                n_features=n_feats,
                k_folds=k_folds,
                random_state=random_state,
                save_dir_eval=RETRAIN_EVAL_DIR
            )

            # (h) Fit full + export ONNX
            onnx_path = fit_full_and_export_onnx_regression(
                model_name=model_name,
                estimator_or_pipeline=model_to_use,
                X=X_sub,
                y=y_train,
                n_features=n_feats,
                save_dir_onnx=RETRAIN_ONNX_DIR
            )

            summary["ONNXPath"] = onnx_path
            results.append(summary)

    return pd.DataFrame(results)
```

```python
# Target Folder Penyimpanan hasil
RETRAIN_ROOT = f"{ROOT_FOLDER}/Regresi/Retraining_Report"
RETRAIN_EVAL_DIR = f"{RETRAIN_ROOT}/Evaluation"
RETRAIN_ONNX_DIR = f"{RETRAIN_ROOT}/ONNX_Models"

# Running
X_train = df_train_reg.drop(columns="Sisa_Usia_Jam")
y_train = df_train_reg["Sisa_Usia_Jam"]

# Folder SHAP Report
SHAP_DIR = f"{ROOT_FOLDER}/Regresi/SHAP_Report"

retrain_results_df = retrain_models_with_shap_features_regression(
    models=models,
    X_train=X_train,
    y_train=y_train,
    shap_dir=SHAP_DIR,
    feature_range=range(3, 17),
    k_folds=10,
    random_state=42
)

# Simpan ringkasan semua hasil retraining
summary_path = f"{RETRAIN_ROOT}/Retraining_Summary_3-16_Features.csv"
retrain_results_df.to_csv(summary_path, index=False)

print("\n[OK] Retraining summary saved:", summary_path)
retrain_results_df
```
